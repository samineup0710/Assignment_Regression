```python
print('We are using part B with 3 layer and 10 nodes')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
concrete_data.head()
```

```python
concrete_data.isnull().sum()
```

```python
concrete_data_columns = concrete_data.columns
```

```python
X = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
Y = concrete_data['Strength'] # Strength column
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X, Y, test_size = 0.3) 
```

```python
X.describe()
```

```python
X_train.describe()
```

```python
predictors_train_norm = ((X_train - X_train.mean())/X_train.std())
predictors_test_norm = ((X_test - X_test.mean())/X_test.std())
```

```python
target_train_norm = ((Y_train - Y_train.mean())/Y_train.std())
target_test_norm = ((Y_test - Y_test.mean())/Y_test.std())
```

```python
n_cols = predictors_train_norm.shape[1]
```

```python
n_cols
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
import tensorflow as tf
```

```python
def regression_model():
    model =keras.Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(10, activation= 'relu'))
    model.add(Dense(10, activation= 'relu'))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
model = regression_model()
model.fit(predictors_train_norm, target_train_norm, validation_split=0.3, epochs =50, verbose=1) #now changing the epoch to 100 for partC
```

```python
from sklearn import metrics
import numpy as np
target_pred = model.predict(predictors_test_norm)
mserrD = metrics.mean_squared_error(target_test_norm, target_pred)
r2_scoresD = metrics.r2_scores = metrics.r2_score(target_test_norm, target_pred)
print('mserrA: {:.2f}'.format(mserrD))
print('mse_StdDev: {:.2f}'.format(r2_scoresD))
print('R^2_Mean: {:.2f}'.format(np.mean(r2_scoresD)))
print('R^2_StdDev: {:.2f}'.format(np.std(r2_scoresD)))
```
